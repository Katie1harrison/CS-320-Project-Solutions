# Fitting with `np.linalg.solve`

**Above:** we estimated house prices using a linear model based on the dot product as follows:

$Xc = y$

* $X$ (known) is a matrix with house features (from DataFrame)
* $c$ (known) is a vector of coefficients (our model parameters)
* $y$ (computed) are the prices

**Below:** what if X and y are know, and we want to find c?

```python
houses = pd.DataFrame([[2,1,1985,196.55],
                       [3,1,1998,260.56],
                       [4,3,2005,334.55],
                       [4,2,2020,349.60]],
                      columns=["beds", "baths", "year", "price"])
houses["one"] = 1
houses
```

If we assume price is linearly based on the features, with this equation:

* $beds*c_0 + baths*c_1 + year*c_2 + 1*c_3 = price$

Then we get four equations:

* $2*c_0 + 1*c_1 + 1985*c_2 + 1*c_3 = 196.55$
* $3*c_0 + 1*c_1 + 1998*c_2 + 1*c_3 = 260.56$
* $4*c_0 + 3*c_1 + 2005*c_2 + 1*c_3 = 334.55$
* $4*c_0 + 2*c_1 + 2020*c_2 + 1*c_3 = 349.60$


```python
X = houses[["beds", "baths", "year", "one"]].values
X
```

```python
y = houses[["price"]].values
y
```

```python
c = np.linalg.solve(X, y)
c
```

# Two Perspectives on `Matrix @ vector`

$\begin{bmatrix}

```python
X = np.array([[4,5], [6,7], [8,9]])
c = np.array([2, 3]).reshape(-1,1)
X @ c
```

## Row Picture

Do dot product one row at a time.

$\begin{bmatrix}
4&5\\6&7\\8&9\\
\end{bmatrix}
\cdot
\begin{bmatrix}
2\\3\\
\end{bmatrix}
=
\begin{bmatrix}
(4*2)+(5*3)\\
(6*2)+(7*3)\\
(8*2)+(9*3)\\
\end{bmatrix}
=
\begin{bmatrix}
23\\
33\\
43\\
\end{bmatrix}
$

```python
def row_dot_product(X, c):
    rv = []
    # loop over rows of X
    for row_idx in range(X.shape[0]):
        row = X[row_idx:row_idx+1, :]
        # dot product each row of our matrix with c
        rv.append((row @ c).item())
    return np.array(rv).reshape(-1,1)
    
row_dot_product(X, c)
```

## Column Picture

$\begin{bmatrix}
c_0&c_1&c_2\\
\end{bmatrix}
\cdot
\begin{bmatrix}
x\\y\\z\\
\end{bmatrix}
=(c_0*x) + (c_1*y) + (c_2*z)
$

Dot product takes a **linear combination** of columns.

$\begin{bmatrix}
4&5\\6&7\\8&9\\
\end{bmatrix}
\cdot
\begin{bmatrix}
2\\3\\
\end{bmatrix}
=
\begin{bmatrix}
4\\6\\8\\
\end{bmatrix}*2
+
\begin{bmatrix}
5\\7\\9\\
\end{bmatrix}*3
=
\begin{bmatrix}
23\\
33\\
43\\
\end{bmatrix}
$

```python
# same result as before, but different definition/code
def col_dot_product(X, c):
    result = np.zeros(X.shape[0]).reshape(-1,1)
    # loop over the columns of X, multiply them, and to the result column
    for col_idx in range(X.shape[1]):
        col = X[:, col_idx:col_idx+1]
        weight = c[col_idx, 0]
        # print(col)
        # print(weight)
        # print(col * weight)
        result += col * weight
    
    return result
    
col_dot_product(X, c)
```
