## Where do numpy arrays show up in ML?

```python
import pandas as pd
from sklearn.linear_model import LinearRegression
```

```python
df = pd.DataFrame([[0,2,1], [2,3,4], [8,5,6]], columns=["x1", "x2", "y"])
df
```

```python
print(type(df.values))
df.values
```

```python
model = LinearRegression()
model.fit(df[["x1", "x2"]], df["y"])
model.coef_
```

```python
model.predict(df[["x1", "x2"]])
```

```python
df[["x1", "x2"]].values @ model.coef_ + model.intercept_
```

## How to slice numpy arrays?

```python
# REVIEW: Python slicing of lists
a = [7,8,9,10]
b = a[1 :  : 2]  # WANT 8 and 10; slicing a list creates a brand new list with a copy of all values
b
```

```python
b[1] = 100
b
```

```python
a # did not change
```

```python
# slicing of numpy arrays
a = np.array([7,8,9,10])
b = a[1 :  : 2]  # WANT 8 and 10; slicing an array shares the values
b
```

```python
b[1] = 100
b
```

```python
a # A changes too!
```

```python
a = [1,2,3]
b = a.copy() # copy everything instead of sharing
b[0] = 99
a
```

## Vector Multiplication: Overview

### Elementwise Multiplication

$\begin{bmatrix}
1 \\ 2 \\ 3
\end{bmatrix}
*
\begin{bmatrix}
4 \\ 5 \\ 6
\end{bmatrix}$

$\begin{bmatrix}
1 \\ 2 \\ 3
\end{bmatrix}
*
\begin{bmatrix}
4 & 5 & 6
\end{bmatrix}$

### Dot Product

$\begin{bmatrix}
1 & 2 & 3
\end{bmatrix}
\cdot
\begin{bmatrix}
4 \\ 5 \\ 6
\end{bmatrix}$

$\begin{bmatrix}
1 \\ 2 \\ 3
\end{bmatrix}
\cdot
\begin{bmatrix}
4 & 5 & 6
\end{bmatrix}$

```python
v1 = np.array([1,2,3]).reshape(-1, 1)
v1
```

```python
v2 = np.array([4,5,6]).reshape(-1, 1)
v2
```

```python
v1 * v2   # [1*4, 2*5, 3*6]
```
