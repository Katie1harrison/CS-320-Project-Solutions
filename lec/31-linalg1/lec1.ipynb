## How to slice numpy arrays?

```python
# review lists
a = [7,8,9,10]
b = a[1 :  : 2]  # want 8 and 10; when slicing a list, we get totally new list object
b
```

```python
b[1] = 100
b
```

```python
a
```

```python
# slicing with numpy
a = np.array([7,8,9,10])
b = a[1 :  : 2]  # want 8 and 10; when slicing a numpy, memory for the numbers is shared
b
```

```python
b[1] = 100
b
```

```python
a # NOT like a list
```

```python
a = np.array([1,2,3])
b = a.copy() # don't share internal values anymore
b[0] = 99
a
```

```python
b
```

## Vector Multiplication: Overview

### Elementwise Multiplication

$\begin{bmatrix}
1 \\ 2 \\ 3
\end{bmatrix}
*
\begin{bmatrix}
4 \\ 5 \\ 6
\end{bmatrix}$

$\begin{bmatrix}
1 \\ 2 \\ 3
\end{bmatrix}
*
\begin{bmatrix}
4 & 5 & 6
\end{bmatrix}$

### Dot Product

$\begin{bmatrix}
1 & 2 & 3
\end{bmatrix}
\cdot
\begin{bmatrix}
4 \\ 5 \\ 6
\end{bmatrix}$

$\begin{bmatrix}
1 \\ 2 \\ 3
\end{bmatrix}
\cdot
\begin{bmatrix}
4 & 5 & 6
\end{bmatrix}$

```python
v1 = np.array([1,2,3]).reshape(-1, 1)
v1
```

```python
v2 = np.array([4,5,6]).reshape(-1, 1)
v2
```

```python
v1 * v2   # [1*4, 2*5, 3*6]
```
