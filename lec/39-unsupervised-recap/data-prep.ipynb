```python
import pandas as pd
```

```python
df = pd.read_csv("Eco-Totem_Capital_City_Trail_Bike_Counts.csv")
df["Count_Date"] = pd.to_datetime(pd.to_datetime(df["Count_Date"]).dt.date)
df.head()
```

```python
counts = df.groupby(by="Count_Date")[["Count"]].sum().reset_index()
counts.head()
```

```python
calendar = counts["Count_Date"].dt.isocalendar().astype(str)
calendar.head()
```

```python
weeks = pd.DataFrame({
    "week": calendar["year"]+"-"+calendar["week"],
    "day": calendar["day"],
    "count": counts["Count"]
}).pivot(index="week", columns="day")
weeks.head()
```

```python
weeks = weeks[~weeks.isna().any(axis=1)]
weeks.head()
```

```python
weeks["count"].to_csv("bikes.csv")
```
